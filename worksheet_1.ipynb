Problem - 1: Array Creation:

```python
import numpy as np

# 1. Empty array 2x2
empty_array = np.empty((2, 2))
print("Empty Array:\n", empty_array)

# 2. All ones array 4x2
ones_array = np.ones((4, 2))
print("\nAll Ones Array:\n", ones_array)

# 3. Full array of given shape and value
filled_array = np.full((3, 3), 7, dtype=int)
print("\nFilled Array:\n", filled_array)

# 4. Zeros like given array
original_array = np.array([[5, 6], [7, 8]])
zeros_like_array = np.zeros_like(original_array)
print("\nZeros Like Array:\n", zeros_like_array)

# 5. Ones like given array
ones_like_array = np.ones_like(original_array)
print("\nOnes Like Array:\n", ones_like_array)

# 6. Convert list to numpy array
new_list = [1, 2, 3, 4]
array_from_list = np.array(new_list)
print("\nArray from List:\n", array_from_list)
```

Problem - 2: Array Manipulation: Numerical Ranges and Array indexing:

```python
import numpy as np

# 1. Array 10 to 49
arr1 = np.arange(10, 50)
print("Array 10 to 49:\n", arr1)

# 2. 3x3 matrix 0 to 8
arr2 = np.arange(9).reshape((3, 3))
print("\n3x3 Matrix 0 to 8:\n", arr2)

# 3. 3x3 identity matrix
identity_matrix = np.eye(3)
print("\n3x3 Identity Matrix:\n", identity_matrix)

# 4. Random array of size 30 and mean
rand_array = np.random.random(30)
mean_val = rand_array.mean()
print("\nRandom Array:\n", rand_array)
print("Mean:", mean_val)

# 5. 10x10 random array, min & max
rand_10x10 = np.random.random((10, 10))
min_val = rand_10x10.min()
max_val = rand_10x10.max()
print("\n10x10 Random Array:\n", rand_10x10)
print("Min:", min_val, "Max:", max_val)

# 6. Zero array of size 10, 5th element 1
zero_array = np.zeros(10)
zero_array[4] = 1
print("\nZero Array with 5th element 1:\n", zero_array)

# 7. Reverse an array
arr = np.array([1, 2, 0, 0, 4, 0])
reversed_arr = arr[::-1]
print("\nReversed Array:\n", reversed_arr)

# 8. 2D array with 1 on border and 0 inside
border_array = np.ones((5, 5))
border_array[1:-1, 1:-1] = 0
print("\n2D Array with border 1 and inside 0:\n", border_array)

# 9. 8x8 checkerboard pattern
checkerboard = np.zeros((8, 8), dtype=int)
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("\n8x8 Checkerboard Pattern:\n", checkerboard)
```

Problem - 3: Array Operations:

```python
import numpy as np

# Given arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# 1. Add the two arrays
add_xy = x + y
print("1. x + y:\n", add_xy)

# 2. Subtract the two arrays
sub_xy = x - y
print("\n2. x - y:\n", sub_xy)

# 3. Multiply the array with an integer
mul_x = x * 3
print("\n3. x * 3:\n", mul_x)

# 4. Square of each element
square_x = x**2
print("\n4. Square of x:\n", square_x)

# 5a. Dot product of v and w
dot_vw = np.dot(v, w)
print("\n5a. Dot product v . w:", dot_vw)

# 5b. Dot product of x and v
dot_xv = np.dot(x, v)
print("\n5b. Dot product x . v:\n", dot_xv)

# 5c. Dot product of x and y
dot_xy = np.dot(x, y)
print("\n5c. Dot product x . y:\n", dot_xy)

# 6a. Concatenate x and y along rows
concat_xy_row = np.concatenate((x, y), axis=0)
print("\n6a. Concatenate x and y along rows:\n", concat_xy_row)

# 6b. Concatenate v and w along columns
concat_vw_col = np.column_stack((v, w))
print("\n6b. Concatenate v and w along columns:\n", concat_vw_col)

# 7. Concatenate x and v
try:
    concat_xv = np.concatenate((x, v), axis=0)
except ValueError as e:
    print("\n7. Error when concatenating x and v:", e)

# Explanation:
# x has shape (2,2) and v has shape (2,), so np.concatenate fails because
# the number of dimensions along axis=0 do not match.

# Fix by reshaping v to 2D (row vector)
v_reshaped = v.reshape(1, 2)
concat_xv_fixed = np.concatenate((x, v_reshaped), axis=0)
print("\n7. Concatenate x and reshaped v:\n", concat_xv_fixed)
```

```python
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])
B_eq = np.array([-1, -3, 9])

# Using inverse method
A_inv_eq = np.linalg.inv(A_eq)
X = np.dot(A_inv_eq, B_eq)
print("\nSolution using inverse method: [x, y, z] =", X)

# Using np.linalg.solve
X_solve = np.linalg.solve(A_eq, B_eq)
print("Solution using np.linalg.solve: [x, y, z] =", X_solve)
```
