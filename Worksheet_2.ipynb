# Problem 2 - Data Imputations:

1. Load the provided dataset and import in pandas DataFrame.

```python
df = pd.read_csv("medical_students_dataset.csv")
df
```


2. Check info of the DataFrame and identify column with missing (null) values.

```python
df.isnull().sum()
```

3. For the column with missing values fill the values using various techniques we discussed above. Try to explain why did you select the particular methods for particular column. italicized text

a. Mean because These measurements are usually normally distributed.

```python
df["Height"] = df["Height"].fillna(df["Height"].mean());
df["Weight"] = df["Weight"].fillna(df["Weight"].mean());
df["Heart Rate"] = pd.to_numeric(df["Heart Rate"], errors="coerce")
df["Heart Rate"] = df["Heart Rate"].fillna(df["Heart Rate"].median());
```

b. Health-related values often contain outliers (e.g., very high BP or cholesterol). so we use median

```python
from numpy import median

df["BMI"] = df["BMI"].fillna(df["BMI"]).median();
df["Blood Pressure"] = df["Blood Pressure"].fillna(df["Blood Pressure"]).median();
df ["Cholesterol"] = df["Cholesterol"].fillna(df["Cholesterol"]).median();
df["Age"] = df["Age"].fillna(df["Age"].median())

```


c. Categorical values cannot use mean/median.:

```python
df["Smoking"] = df ["Smoking"].fillna(df["Smoking"].mode()[0]);
df["Diabetes"] = df["Diabetes"].fillna(df["Diabetes"].mode()[0]);
df["Blood Type"] = df["Blood Type"].fillna(df["Blood Type"].mode()[0])
```

d. Student Id should be unique soo

```python

df["Student ID"] = df.index + 1
```

# Data Cleaning and Transformations

Problem - 1:
Create a DataFrame that is subsetted for the columns ’Name’, ’Pclass’, ’Sex’, ’Age’, ’Fare’, and ’Survived’.
Retain only those rows where ’Pclass’ is equal to 1, representing first-class passengers. What is the mean,
median, maximum value, and minimum value of the ’Fare’ column?


```python
import pandas as pd;
df = pd.read_csv("Titanic-Dataset.csv")
subset = df[["Name","Pclass","Sex","Age","Fare","Survived"]]
first_class = subset[subset["Pclass"]==1]
mean_fare = first_class["Fare"].mean()
median_fare = first_class["Fare"].median()
maximum_value = first_class["Fare"].max()
minimum_value = first_class["Fare"].min()
print("The mean fare is",mean_fare)
print("The median fare is",median_fare)
print("The maximum fare is",maximum_value)
print("The minimum fare is",minimum_value)
```

Problem - 4:

Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column. Draw a visual-
ization to show how the survival distributions vary by gender.

```python
# Mean survival rate by gender
mean_survival_by_sex = df.groupby('Sex')['Survived'].mean()
print(mean_survival_by_sex)
import matplotlib.pyplot as plt

# Plot survival rate by gender
mean_survival_by_sex.plot(kind='bar')

plt.title("Survival Rate by Gender")
plt.xlabel("Sex")
plt.ylabel("Mean Survival Rate")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

Problem - 5:

Draw a visualization that breaks your visualization from Exercise 3 down by the port of embarkation (’Em- barked’). In this instance, compare the ports ’C’ (Cherbourg), ’Q’ (Queenstown), and ’S’ (Southampton). italicized text

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = sns.load_dataset("titanic")

# Group survival rate by embarked port
embarked_survival = df.groupby('embarked')['survived'].mean()
print(embarked_survival)

# Visualization
plt.figure(figsize=(6,4))
plt.bar(embarked_survival.index, embarked_survival.values)
plt.xlabel("Embarked Port (C, Q, S)")
plt.ylabel("Mean Survival Rate")
plt.title("Survival Rate by Embarkation Port")
plt.show()
```
