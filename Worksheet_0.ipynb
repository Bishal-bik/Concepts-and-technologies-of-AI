Task 1 - Sum of Nested Lists:

```python
nested_list = [10, [20, 30], [40, [50, [60]]]]

def sum_nested_list(nested_list):
   total = 0

   for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)  # recursion for nested list
        else:
            total += element  # add number directly
   return total

print("Nested List:", nested_list)
print("Total Sum of nested list:", sum_nested_list(nested_list))
```

Task 2 - Generate All Permutations of a String:

```python
def generate_permutations(s):
    """
    Generate all unique permutations of a given string using recursion.
    Returns a list of strings.
    """
    # Base case: if string has 1 or 0 characters, return it as the only permutation
    if len(s) <= 1:
        return [s]

    permutations = set()   # use a set to avoid duplicates

    # Recursive case
    for i, char in enumerate(s):
        remaining = s[:i] + s[i+1:]   # remove the chosen character
        for perm in generate_permutations(remaining):
            permutations.add(char + perm)

    return list(permutations)
print("Permutations of (abc) are:",generate_permutations("abc"))
print("Permutations of (aab) are:",generate_permutations("aab"))




```

Task 3 - Directory Size Calculation:

```python
def calculate_directory_size(directory):

    total_size = 0

    for name, content in directory.items():
        if isinstance(content, dict):  # if it's a subdirectory
            total_size += calculate_directory_size(content)  # recursive call
        else:  # it's a file, add its size
            total_size += content

    return total_size

# Sample directory structure
directory_structure = {
"file1.txt": 200,
"file2.txt": 300,
"subdir1": {
"file3.txt": 400,
"file4.txt": 100
},
"subdir2": {
"subsubdir1": {
"file5.txt": 250
},
"file6.txt": 150
}
}

total = calculate_directory_size(directory_structure)
print("Total directory size:", total, "KB")
```

8.2.1 Exercises - Dynamic Programming:

Task 1 - Coin Change Problem:

```python
def min_coins(coins, amount):

    # Initialize DP array; dp[i] = minimum coins needed for amount i
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins needed to make amount 0

    # Build up the DP table
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    # If amount is still infinity, it means it's not possible
    return dp[amount] if dp[amount] != float('inf') else -1

coins = [1, 2, 5]
amount = 11

result = min_coins(coins, amount)
print("Minimum coins needed:", result)
```

Task 2 - Longest Common Subsequence (LCS):

```python
def longest_common_subsequence(s1, s2):
    """
    Finds the length of the Longest Common Subsequence (LCS) between two strings.

    Args:
        s1 (str): First string.
        s2 (str): Second string.

    Returns:
        int: Length of the LCS.

    Example:
        >>> longest_common_subsequence("abcde", "ace")
        3
    """
    m, n = len(s1), len(s2)

    # Create DP table: dp[i][j] = LCS length for s1[0..i-1] & s2[0..j-1]
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Fill the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]

s1 = "abcde"
s2 = "ace"

lcs_length = longest_common_subsequence(s1, s2)
print("Length of LCS:", lcs_length)
```

Task 3 - 0/1 Knapsack Problem:

```python
def knapsack(weights, values, capacity):
    """
    Determines the maximum value of items that can be carried within a given weight capacity.

    Args:
        weights (list of int): List of item weights.
        values (list of int): List of item values.
        capacity (int): Maximum weight capacity of the knapsack.

    Returns:
        int: Maximum total value achievable without exceeding the capacity.

    Example:
        >>> knapsack([1, 3, 4, 5], [1, 4, 5, 7], 7)
        9
    """
    n = len(weights)
    # Create DP table: dp[i][w] = max value using first i items with weight limit w
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Fill DP table
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]


weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

max_value = knapsack(weights, values, capacity)
print("Maximum value achievable:", max_value)
```
